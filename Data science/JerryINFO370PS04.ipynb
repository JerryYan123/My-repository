```python
#1.4
import numpy as np
import matplotlib.pyplot as plt

def my_function(R, S):
    result = []
    for i in range(R):
        X = np.random.randint(0, 2, size=S)*2 - 1
        result.append(np.mean(X))
    plt.hist(result, bins=100)
    plt.show()
    return result

result = my_function(1000, 1)
print("The historgram has a binomial shape, where around half of them distributed at -1 and the rest at 1.")
```

```python
#1.5
print("mean of the sample:", np.mean(result))
print("variance of the sample:", np.var(result))
```

```python
#2.2.4

p2sample = []
for i in range(R):
    X = np.random.binomial(1, p2, size=S)
    p2sample.append(np.mean(X))
```

```python
#2.2.5

print("sample mean:", np.mean(p2sample))
print("The average of the averages is resemble to p2, 0.51. According to the Law of Large Number, the larger the sample size, the closer the mean to the true mean, which is p2.")
```

```python
#2.2.6

plt.hist(p2sample, bins=100)
plt.show()
print("The distribution resemble to 0.51")
print("The smallest value is around 0.445, the largest value is around 0.57, according to eyeballing the value that is reasonably common")
```
